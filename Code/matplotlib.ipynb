```python
x = np.arange(0, 4 * np.pi, .1) 
y = np.cos(x) 
plt.title("sine wave form") 
plt.plot(x, y,c='r') 
plt.show()
```

```python
plt.subplot(3,1,1) #(row,col,index)
plt.plot(x,y,c='r')
plt.title('Graph1')
plt.subplot(3,1,2) #(row,col,index)
plt.plot(x,y,c='g')
plt.title('Graph2')
plt.subplot(3,1,3) #(row,col,index)
plt.plot(x,y,c='b')
plt.title('Graph3')

```
